```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pathlib
from sklearn.ensemble import RandomForestClassifier
import os
```

```python

n = 100  # Pilih 100 sampel pelatihan
np.random.seed(100)  # Seed agar hasil acak bisa direproduksi

# Untuk menyimpan fitur dan label
fea_lst = []
y_lst = []

# Daftar file dataset
files = ['1.csv', '2.csv', '3.csv', '4.csv', '5.csv', '6.csv', '7.csv', '8.csv']

# Iterasi untuk setiap file
for i, f in enumerate(files):
    print(i, f)

    if not os.path.exists(f):
        print(f"File '{f}' tidak ditemukan, dilewati.")
        continue

    # Membaca file
    ya = pd.read_csv(f)

    # Mengisi missing value
    ya.ffill(inplace=True)
    ya.bfill(inplace=True)

    # Sampling n baris (atau kurang kalau datanya sedikit)
    data = ya.values
    idx = np.random.choice(data.shape[0], min(n, data.shape[0]), replace=False)
    arr = data[idx, :]

    # Simpan fitur dan label
    fea_lst.append(arr)
    y_lst.append(np.full(arr.shape[0], i))

    print(f'ya.shape: {ya.shape}, dipilih: {arr.shape[0]} sampel')

# Menggabungkan semua data
fea_name = list(ya.columns)  # Ambil nama fitur dari file terakhir
x_data = np.vstack(fea_lst)
y_label = np.hstack(y_lst)

print("Label:", y_label)
print(f'x_data shape: {x_data.shape}')
print(f'y_label shape: {y_label.shape}')

# Inisialisasi Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(x_data, y_label)

# Evaluasi akurasi
print(f'Akurasi Model: {clf.score(x_data, y_label)}')

# Menentukan pentingnya fitur
fea_dict = {fea_name[i]: clf.feature_importances_[i] for i in range(len(fea_name))}
fea_import = sorted([(v, k) for k, v in fea_dict.items()], reverse=True)

print('Pentingnya Fitur yang Telah Diurutkan:')
for rank, (imp, feat) in enumerate(fea_import, 1):
    print(f"{rank}. {feat} - Importance: {imp:.4f}")

# Simpan ke folder
header = "Feature,Importance"
out_dir = r'C:\Users\MANDA\Documents\SEMESTER 6\output_hw7'
if not os.path.exists(out_dir):
    os.makedirs(out_dir)

# Simpan hasil fitur penting ke satu file CSV
output_csv = os.path.join(out_dir, 'feature_importances.csv')
with open(output_csv, 'w') as f:
    f.write(f"{header}\n")
    for importance, feature in fea_import:
        f.write(f"{feature},{importance}\n")

print(f"Hasil fitur penting disimpan dalam: {output_csv}")


```
