```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
# Baca data dari file CSV (kolom 1 sampai 4 saja, sebagai integer)
data = np.genfromtxt('1.csv', delimiter=',', usecols=(1, 2, 3, 4), dtype='int32')

# Ambil subset data pada baris ke-1000 sampai 1099, kolom ke-2 (index 1)
subset = data[1000:1100, 1]

# Cari puncak dengan jarak minimum antar puncak 10 dan 3
peaks_10, _ = find_peaks(subset, distance=10)
peaks_3, _ = find_peaks(subset, distance=3)

# Cetak hasil indeks puncak
print('Puncak (distance=10):', peaks_10)
print('Puncak (distance=3):', peaks_3)

# Visualisasi data dan puncaknya
plt.figure(figsize=(12, 5))
plt.plot(subset, label='Data')
plt.plot(peaks_3, subset[peaks_3], 'vr', label='Puncak (distance=3)')
plt.plot(peaks_10, subset[peaks_10], 'xk', label='Puncak (distance=10)')
plt.title('Deteksi Puncak dengan Jarak Minimum 3 dan 10')
plt.xlabel('Indeks')
plt.ylabel('Nilai')
plt.legend()
plt.grid(True)
plt.show()
```
